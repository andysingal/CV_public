Qunatize Model

```python
from transformers import BitsAndBytesConfig
import torch

bnb_config = BitsAndBytesConfig(
    load_in_8bit=True
)
```

```python
import torch
from PIL import Image
from transformers import AutoModel, AutoTokenizer

model = AutoModel.from_pretrained('openbmb/MiniCPM-V-2_6',
                                   trust_remote_code=True,
                                   attn_implementation='eager',
                                   torch_dtype=torch.bfloat16,
                                   device_map = "cuda",
                                   quantization_config=bnb_config)

model = model.eval()

tokenizer = AutoTokenizer.from_pretrained('openbmb/MiniCPM-V-2_6', trust_remote_code=True)
```

Signature verification

```python
image1 = Image.open('sign_a.png').convert('RGB')
image2 = Image.open('sign_b.png').convert('RGB')
question = 'Provided two signatures, find out if they are of the same person or not, just answere with "match" or "no match" nothing else.'

msgs = [{'role': 'user', 'content': [image1, image2, question]}]

answer = model.chat(
    image=None,
    msgs=msgs,
    tokenizer=tokenizer
)
print(answer)
```

```python
import os
files = os.listdir("./")
lst = []
for f in files:
  if f.endswith(".png"):
    lst.append(f)
lst
```

```python
for img in lst:
  for img2 in lst:
    image1 = Image.open(img).convert('RGB')
    image2 = Image.open(img2).convert('RGB')
    msgs = [{'role': 'user', 'content': [image1, image2, question]}]

    answer = model.chat(
        image=None,
        msgs=msgs,
        tokenizer=tokenizer
    )
    print(img, img2, answer)
```
