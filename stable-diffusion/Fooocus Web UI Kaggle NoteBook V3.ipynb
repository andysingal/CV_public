```python
# Download all generated images. To do that make sure to cancel above running cell of Fooocus Web UI
import os
import zipfile

# Define the directory to be zipped
source_dir = '/kaggle/working/Fooocus/outputs'

# Define the name of the output zip file
zip_filename = '/kaggle/working/Kaggle_Fooocus_Images.zip'

# Check if the zip file already exists and delete it if it does
if os.path.exists(zip_filename):
    os.remove(zip_filename)
    print(f'Deleted existing file: {zip_filename}')

# Create a zip file
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            # Create a relative path for files to keep the directory structure
            relative_path = os.path.relpath(os.path.join(root, file), os.path.join(source_dir, '..'))
            zipf.write(os.path.join(root, file), relative_path)

print(f'Zip file created: {zip_filename}')
```
